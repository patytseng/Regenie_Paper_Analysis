```r
library(ggplot2)
library(ggpubr)
```

```r
pheno <- read.csv('~/Torkamani_Projects/20210712_Analysis-RegeniePaper/CAD_PRS/PRS_validator/analysis/aric_cad_post_stdzn_prs_pheno.csv')
pheno$ID <- as.character(pheno$SAMPLE_ID)
pheno <- pheno[,c('ID','ISP12','AGISP12')]
```

```r
int <- 1
scores <- read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_1-N_8264-chrALL.csv'))
scores <- rbind(scores,read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_3-N_3-chrALL.csv')))
scores <- rbind(scores,read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_5-N_508-chrALL.csv')))
scores <- rbind(scores,read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_mixed-N_4284-chrALL.csv')))
scores$ID <- sapply(strsplit(scores$X.sampleID,'_'),'[[',2)
scores$PRS <- scores$sum_PRS / scores$SNP_N
scores <- scores[,c('ID','PRS')]
names(scores) <- c('ID',paste0('V',int,'_PRS'))

for (int in 2:30){
    file <- read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_1-N_8264-chrALL.csv'))
    file <- rbind(file,read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_3-N_3-chrALL.csv')))
    file <- rbind(file,read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_5-N_508-chrALL.csv')))
    file <- rbind(file,read.csv(paste0('210729-V',int,'.tmp-PRS_1047343-aric_c1-ancestry_mixed-N_4284-chrALL.csv')))
    file$ID <- sapply(strsplit(file$X.sampleID,'_'),'[[',2)
    file$PRS <- file$sum_PRS / file$SNP_N
    file <- file[,c('ID','PRS')]
    names(file) <- c('ID',paste0('V',int,'_PRS'))
    scores <- merge(scores,file,by='ID')
}

data <- merge(scores, pheno,by='ID')
data$ISP12 <- as.factor(data$ISP12)
```

```r
for (i in 1:30) {
    data$std <- ( ( data[[i+1]] - mean(data[[i+1]] )) / sd(data[[i+1]]) )
    colnames(data)[which(names(data) == "std")] <- paste0('V',i,'_PRS_std')
}
```

```r
V_names = glue::glue('V{i}_PRS_std',i=1:30)
plots <- vector('list', 30)

for (i in 1:30) {
    plots[[i]] <- local({
        i <- i
        plot <- ggplot(data, aes_string(x = V_names[[i]], color = 'ISP12')) + 
        geom_density() + 
        geom_density(color = 'black') + 
        geom_vline(xintercept = 0, linetype = 'longdash') +
        theme_classic()
    })
}
```

```r
ggarrange(plots[[1]],plots[[2]],plots[[3]],plots[[4]],plots[[5]],
          plots[[6]],plots[[7]],plots[[8]],plots[[9]],plots[[10]],
          plots[[11]],plots[[12]],plots[[13]],plots[[14]],plots[[15]],
          plots[[16]],plots[[17]],plots[[18]],plots[[19]],plots[[20]],
          plots[[21]],plots[[22]],plots[[23]],plots[[24]],plots[[25]],
          plots[[26]],plots[[27]],plots[[28]],plots[[29]],plots[[30]])
ggsave('PRS_Plots.png', device = 'png', width =30, height =25, units = "in",
  path = '~/Torkamani_Projects/20210712_Analysis-RegeniePaper/CAD_PRS/PRS_Calculated/',)
```
