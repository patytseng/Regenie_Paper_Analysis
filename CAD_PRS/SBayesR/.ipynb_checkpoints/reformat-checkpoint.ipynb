```python
import pandas as pd
import numpy as np
```

```python
weights = []

for i in range(0,22):
    weights.append(pd.read_csv(f'~/Torkamani_Projects/20210712_Analysis-RegeniePaper/CAD_PRS/SBayesR/exBTs_regenie_phenoCol1_SPA_CADComp.regenie.chr{i+1}.reweighted.snpRes',sep='\s+',engine='python'))
    weights[i] = weights[i][['Chrom','Position','A1','A2','A1Effect']]
    weights[i] = weights[i].rename(columns={'Chrom': 'chr', 'Position': 'bp', 'A1':'non_effect_allele', 'A2':'effect_allele', 'A1Effect':'weight'})
```

```python
weights[0]
```

```python
for i in range(0,22):
    subdata = weights[i].loc[weights[i]['weight'] < 0]
    subdata = pd.DataFrame({
        'chr':subdata['chr'],
        'bp':subdata['bp'],
        'effect_allele':subdata['non_effect_allele'],\
        'non_effect_allele':subdata['effect_allele'],
        'weight':(subdata['weight']*-1)
    })
    weights[i].update(subdata)
    weights[i] = weights[i].astype({'chr': int, 'bp':int})
```

```python
weights[0]
```

```python
output = weights[0].append(weights[1:])
```

```python
output
```

```python
output.to_csv('exBTs_regenie_phenoCol1_SPA_CADComp.regenie.reweighted.csv', index=False)
```
